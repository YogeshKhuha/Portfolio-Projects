```python
#import libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#read the dataset
fifa= pd.read_csv('players_20.csv')
```

```python
#preview dataset
fifa.head()
```

```python
#total number of rows and columns
fifa.shape
```

```python
# all the column names 
for col in fifa.columns:
    print(col)
```

### Number of Players from each nationality

```python
fifa['nationality'].value_counts()
```

### What are 10 nationalities with the most number of players?

```python
fifa['nationality'].value_counts()[:10]
```

```python
#list of the top 10 nationalities with most number of players
fifa['nationality'].value_counts()[:10].keys()
```

```python
#bar chart for top 10 nationalities with most number of players
plt.figure(figsize=(8,5))
plt.bar(fifa['nationality'].value_counts()[:10].keys(),fifa['nationality'].value_counts()[:10],color='g')
plt.show()
```

### Who are the players who earn the most?

```python
#store the player name and wage in euros in salary
salary=fifa[['short_name','wage_eur']]
```

```python
#preview of data in salary
salary.head()
```

```python
#sorting data in salary by wage in descending order
salary= salary.sort_values(by='wage_eur',ascending=False)
```

```python
#preview salary data after sorting
salary.head()
```

```python
 #bar plot for 5 players with highest wage
plt.figure(figsize=(8,5))
plt.bar(list(salary['short_name'])[:5],list(salary['wage_eur'])[:5],color=['b','g','r','y','orange'])
plt.show()
```

### Analysing players of German nationality with respect to different attributes

```python
#see if nationality of player is germany
fifa['nationality']=='Germany'
```

```python
#store all player info with German nationality in germany 
germany=fifa[fifa['nationality']=='Germany']
```

```python
#display 10 records of germany
germany.head(10)
```

#### Who are the 5 tallest players of German nationality?

```python
germany.sort_values(by=['height_cm'],ascending=False).head()
```

#### Who are the 5 heaviest players of German nationality?

```python
germany.sort_values(by=['weight_kg'],ascending=False).head()
```

#### Who are the 5 highest paid players of German nationality?

```python
germany.sort_values(by=['wage_eur'],ascending=False).head()
```

```python
#display only short name and wage of the highest earning player of german nationality
germany[['short_name','wage_eur']].sort_values(by=['wage_eur'],ascending=False).head()
```

### Which players have the highest shooting skills?

```python
shooting=fifa[['short_name','shooting']]
shooting.sort_values(by=['shooting'],ascending=False).head()
```

### Which players has the highest defending skills?

```python
#defending
defending=fifa[['short_name','defending','nationality','club']]
defending.sort_values(by=['defending'],ascending=False).head()
```
